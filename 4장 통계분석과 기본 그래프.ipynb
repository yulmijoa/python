### 1.7 그래프 위에 텍스트 추가하기

```python
#텍스트 추가 사용자 정의 함수 만들기
def addtext(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]+0.5,y[i],ha='center') 
        #ha파라미터에 center를 지정함으로써 막대의 가로 기준 중앙에 텍스트 표시
```

```python
#막대 위에 텍스트 추가
col7=sns.color_palette('Pastel2',7)
plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'],hat['chick'],color=col7,edgecolor='black')
addtext(hat['hatchery'],hat['chick'])
plt.title('부화장별 병아리 부화현황', fontsize=17) #그래프 이름
plt.xlabel('부화장') #x축 이름
plt.ylabel('부화마릿수') #y축 이름
plt.show()
```

### 1.8 그래프 위에 선 추가하기

```python
col7=sns.color_palette('Pastel2',7)
plt.figure(figsize=(10,7))
plt.bar(hat['hatchery'],hat['chick'],color=col7,edgecolor='black')
addtext(hat['hatchery'],hat['chick'])
plt.title('부화장별 병아리 부화현황', fontsize=17) #그래프 이름
plt.hlines(30, -1, 7, colors='red', linestyles='dashed')
plt.xlabel('부화장') #x축 이름
plt.ylabel('부화마릿수') #y축 이름
plt.show()
```

### 1.9 파이 차트 그려보기

```python
#파이 차트를 그리기 위해 비율 계산
pct=hat['chick']/hat['chick'].sum()
pct
```

```python
#파이 차트 그리기
col7=sns.color_palette('Pastel2',7)
plt.figure(figsize=(10,10))
plt.pie(pct, labels = hat['hatchery'], autopct='%.1f%%', colors=col7, startangle=90, counterclock=False)
#label이름은 hatchery로 지정, autopct는 비율 표시: 첫째자리, False는 시계방향
plt.show()
```

# 3. 가설검정

### 3.1 데이터 불러와서 확인하기

```python
#pandas 불러오기 및 pd라는 약어로 지칭하기
import pandas as pd
```

```python
test=pd.read_csv('ch4-3.csv') #test 변수에 데이터 셋 입력
test
```

### 3.2 상자그림으로 분포 비교하기

```python
#seaborn 패키지 이용 그룹별 상자그림 그리기
import seaborn as sns
plt.figure(figsize=(10,7))
sns.boxplot(x='weight', y='hatchery', data=test)
plt.title('부회장 A vs. B 몸무게 분포 비교', fontsize=17)
plt.show()
```

### 3.3 정규분포인지 검정하기

```python
#가설검정을 위한 패키지 불러오기 및 sp라는 약어로 지칭
import scipy as sp
```

```python
# 부화장 A만 별도로 데이터셋 구성
test_a=test.loc[test.hatchery == 'A', 'weight']
```

```python
# 부화장 B만 별도로 데이터셋 구성
test_b=test.loc[test.hatchery == 'B', 'weight']
```

```python
# 부화장 A 샤피로-윌크 검정 실시
sp.stats.shapiro(test_a)
```

```python
# 부화장 B 샤피로-윌크 검정 실시
sp.stats.shapiro(test_b)
```

### 3.4 t-test로 두 집단 간 평균 검정하기

```python
#부화장 A, B 두 집단 간 평균 검정
sp.stats.ttest_ind(test_a, test_b)
```
