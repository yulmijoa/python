### 결측치 확인하기

```python
import pandas as pd
```

```python
#read_csv()함수 활용 데이터 셋 불러오기
air=pd.read_csv('ch3-1.csv')
air
#결측치는 Nan로 표기됨
```

```python
#결측치 확인하기1
air.info()
```

```python
#결측치 확인하기2   True가 결측치 있음
air.isnull()
```

```python
#결측치 확인하기3    결측치 개수 세어 줌
air.isnull().sum()
```

### 결측치 제거하기

```python
#결측치 제거(axis=0: 행제거, axis=1: 열제거)
air_d=air.dropna(axis=0)
air_d
```

### 결측치를 평균으로 대체하기

```python
#결측치를 평균으로 대체하기 위해 평균 확인
air.mean()
```

```python
#평균 값으로 결측치 대체
air_m=air.fillna(air.mean())
air_m
```

### 결측치를 중앙값으로 대체하기

```python
#중앙값 확인
air.median()
```

```python
air_md=air.fillna(air.median())
air_md
```
